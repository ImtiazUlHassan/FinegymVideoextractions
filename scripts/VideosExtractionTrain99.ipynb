```python
import os
import cv2
import json  
import pandas as pd 

from moviepy import VideoFileClip
import os

def extract_video_clip(videopath, class_name, Subaction_start, Subaction_end, full_id, output_root):
    # Ensure the class_name folder exists
    class_folder = os.path.join(output_root, str(class_name))
    os.makedirs(class_folder, exist_ok=True)

    # Load the video clip
    video = VideoFileClip(videopath)
    
    # Extract the portion of the video
    clip = video.subclipped(Subaction_start, Subaction_end)
    
    # Save the video clip to the class folder with the full_id as the filename
    output_path = os.path.join(class_folder, f"{full_id}.mp4")
    clip.write_videofile(output_path, codec="libx264")

    # Close the video clip to release resources
    video.close()
    return "Done"

# Example usage:
# extract_video_clip('path/to/video.mp4', 1, 5.0, 10.0, '12345', '/output/directory')
```
